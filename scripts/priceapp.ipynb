```python
df_add = pd.read_csv("/Users/harithagb/Downloads/BTC-USD-3.csv")
```

```python
# add 6 new empty columns
df_read["high"] = np.nan
df_read["low"] = np.nan
df_read["open"] = np.nan
df_read["close"] = np.nan
df_read["volume"] = np.nan
df_read["adj_close"] = np.nan
```

```python
# convert date time to date
df_read["date"] = pd.to_datetime(df_read["date"]).dt.date
df_read.head()
```

```python
# convert date time to date
df_add["Date"] = pd.to_datetime(df_add["Date"]).dt.date
df_add.head()
```

```python
type(df_read['date'])
```

```python
date = df_read["date"].iloc[2]
entry = df_add.loc[df_add["Date"] == date]
```

```python
print(date)
```

```python
entry
```

```python
df_read.at[2, "high"] = entry["High"]
df_read.at[2, "low"] = entry["Low"]
df_read.at[2, "open"] = entry["Open"]
df_read.at[2, "close"] = entry["Close"]
df_read.at[2, "volume"] = entry["Volume"]
df_read.at[2, "adj_close"] = entry["Adj Close"]

print(df_read.iloc[2])
```

```python
for ind in tqdm(df_read.index):
    date = df_read["date"].iloc[ind]
    entry = df_add.loc[df_add["Date"] == date]
    df_read.at[ind, "high"] = entry["High"]
    df_read.at[ind, "low"] = entry["Low"]
    df_read.at[ind, "open"] = entry["Open"]
    df_read.at[ind, "close"] = entry["Close"]
    df_read.at[ind, "volume"] = entry["Volume"]
    df_read.at[ind, "adj_close"] = entry["Adj Close"]
```
